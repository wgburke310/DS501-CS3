```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('world-data-2023-clean.csv')
```

```python
df.head()
```

```python
df['density'] = df['Density..P.Km2.'].apply(lambda x: float(x.replace(',','')))
df['land_area_m'] = df['Land.Area.Km2.'].apply(lambda x: float(x.replace(',',''))/1000000)
df['agricultural_land'] = df['Agricultural.Land....'].apply(lambda x: float(x.replace('%',''))/100)
df['birth_rate'] = df['Birth.Rate']
df['fertility_rate'] = df['Fertility.Rate']
df['life_expectancy'] = df['Life.expectancy']
df['co2_emissions'] = df['Co2.Emissions'].apply(lambda x: float(x.replace(',','')))
df['proportion_forested'] = df['Forested.Area....'].apply(lambda x: float(x.replace('%',''))/100)
df['population_m'] = df['Population'].apply(lambda x: float(x.replace(',',''))/1000000)
df['urban_population_m'] = df['Urban_population'].apply(lambda x: float(x.replace(',',''))/1000000)
```

```python
df = df[['Country', 'density','land_area_m','agricultural_land', 'birth_rate','fertility_rate','life_expectancy','proportion_forested','population_m','urban_population_m','co2_emissions']]
df.head()
```

```python
df.to_csv('world-data-2023-clean-2.csv', index=False)
```
